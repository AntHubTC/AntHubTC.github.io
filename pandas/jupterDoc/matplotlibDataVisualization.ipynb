```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print (matplotlib.matplotlib_fname())

# 中文显示问题 参考链接：https://blog.csdn.net/lemonbit/article/details/121433603
# matplotlib其实是不支持显示中文的 显示中文需要一行代码设置字体
matplotlib.rcParams['font.family'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False   # 步骤二（解决坐标轴负数的负号显示问题）
```

```python
# 创建示例 DataFrame
data = {
    '日期': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    '销量': [100, 120, 130, 150, 180, 200, 220, 250, 300, 350]
}
df = pd.DataFrame(data)

# 设置日期为索引
df.set_index('日期', inplace=True)
print(df)

# 绘制折线图
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['销量'], marker='o')
plt.title('每日销量')
plt.xlabel('日期')
plt.ylabel('销量')
plt.grid()
plt.show()
```

## Seaborn

Seaborn 是基于 Matplotlib 的统计数据可视化库，提供了更高级的接口和美观的默认样式。

```python
! pip install seaborn
```

```python
import seaborn as sns

# 使用 Seaborn 绘制折线图
plt.figure(figsize=(10, 5))
sns.lineplot(data=df, x=df.index, y='销量', marker='o')
plt.title('每日销量 - Seaborn')
plt.xlabel('日期')
plt.ylabel('销量')
plt.grid()
plt.show()
```
