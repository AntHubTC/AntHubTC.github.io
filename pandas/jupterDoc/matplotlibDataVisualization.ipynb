```python
import pandas as pd
import matplotlib.pyplot as plt

# 创建示例 DataFrame
data = {
    '日期': pd.date_range(start='2023-01-01', periods=10, freq='D'),
    '销量': [100, 120, 130, 150, 180, 200, 220, 250, 300, 350]
}
df = pd.DataFrame(data)

# 设置日期为索引
df.set_index('日期', inplace=True)

# 绘制折线图
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['销量'], marker='o')
plt.title('每日销量')
plt.xlabel('日期')
plt.ylabel('销量')
plt.grid()
plt.show()
```
